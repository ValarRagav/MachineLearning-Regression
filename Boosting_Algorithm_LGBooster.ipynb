```python
import lightgbm as lgb
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset=pd.get_dummies(dataset,columns=['sex','smoker'],drop_first=True)
#dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#training & test set split
from sklearn.model_selection import train_test_split #import train_test_split from library
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #assign values to 4 variable( train and test set)
```

```python
#print(lightgbm.__version__)
```

```python
regressor = lgb.LGBMRegressor(n_jobs=1, random_state=200,n_estimators=10)

# Train the model
regressor.fit(X_train, Y_train)

```

```python
regressor
```

```python
# Predict
y_pred = regressor.predict(X_test)
```

```python
# Evaluate
#r_score = np.sqrt(mean_squared_error(Y_test, y_pred))
r_score=r2_score(Y_test,y_pred)
r_score
```
