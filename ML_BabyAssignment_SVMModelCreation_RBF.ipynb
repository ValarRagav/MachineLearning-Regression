```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
#dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
#training & test set split
from sklearn.model_selection import train_test_split #import train_test_split from library
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #assign values to 4 variable( train and test set)
```

```python
X_test
```

```python
from sklearn.svm import SVR 
regressor=SVR(kernel="rbf",C=1000) #procedure 
regressor.fit(X_train,Y_train)#Model Creation
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
r_score
```
