```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
#dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
#independent
```

```python
dependent=dataset[['charges']]
#dependent
```

```python
#training & test set split
from sklearn.model_selection import train_test_split #import train_test_split from library
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #assign values to 4 variable( train and test set)
```

```python
X_test
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import make_regression
regressor=AdaBoostRegressor(random_state=0, n_estimators=100,loss='linear')
regressor.fit(X_train,Y_train) 
```

```python
y_pred=regressor.predict(X_test)
#y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
r_score
```
